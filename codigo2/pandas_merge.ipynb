```python
### Unir pedidos, empresas y empleados con merge
```

```python
import pandas as pd
```

```python
dfPedidos = pd.read_csv("../ficheros_curso/merge/Pedidos.txt", sep=';')
dfEmp = pd.read_csv("../ficheros_curso/merge/Empleados.txt", sep=';')
dfEmpresas = pd.read_csv("../ficheros_curso/merge/Empresas.txt", sep=';')
```

```python
dfTotal = pd.merge(dfPedidos, dfEmp, left_on='idempleado', right_on='id')
dfTotal.head()
```

```python
dfTotal = dfTotal.merge(dfEmpresas, left_on='idempresa', right_on='id')
dfTotal.head()
```
