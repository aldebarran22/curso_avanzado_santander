```python
### Unir pedidos, empresas y empleados con merge
```

```python
import pandas as pd
```

```python
dfPedidos = pd.read_csv("../ficheros_curso/merge/Pedidos.txt", sep=';')
dfEmp = pd.read_csv("../ficheros_curso/merge/Empleados.txt", sep=';')
dfEmpresas = pd.read_csv("../ficheros_curso/merge/Empresas.txt", sep=';')
```

```python
dfTotal = pd.merge(dfPedidos, dfEmp, left_on='idempleado', right_on='id')
dfTotal.head()
```

```python
dfTotal = dfTotal.merge(dfEmpresas, left_on='idempresa', right_on='id')
dfTotal.head()
```

```python
dfTotal.drop(columns=['idempleado','idempresa','id_x', 'id_y'], inplace=True)
dfTotal.head()
```

```python
dfTotal.rename(columns={"nombre":"empleado", "cargo":"puesto"}, inplace=True)
dfTotal.head()
```

```python
dfTotal.empresa.value_counts()
```

```python
# Facturación total por empresa:
resul = dfTotal.importe.groupby(dfTotal.empresa).sum()
resul
```

```python
resul2 = dfTotal.importe.groupby(dfTotal.empresa).mean()
resul2
```

```python
resul = dfTotal.importe.groupby([dfTotal.empresa, dfTotal.empleado]).sum()
print(type(resul))
print(resul['Federal Shipping','Buchanan'])
```
