```python
import numpy as np
```

```python
arr= np.arange(10)
arr
```

```python
%%timeit
my_arr = np.arange(1000000)
for _ in range(10): my_arr2 = my_arr * 2
```

```python
%%timeit
my_list = list(range(100000))
for _ in range(10): my_list2 = [x*2 for x in my_list]
```
