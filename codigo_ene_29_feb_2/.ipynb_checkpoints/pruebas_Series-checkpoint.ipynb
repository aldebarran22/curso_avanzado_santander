```python
import pandas as pd
import numpy as np
from pandas import Series
```

```python
s1 = Series([1,4,5,6,9])
s2 = Series(np.arange(10))
```

```python
s3 = Series([1,2,3,4,5], index=list('abcde'))
s3
```

```python
s3['a'], s3[0], s3[-5]
```
