```python
# Indicar el tipo al construir el array
a4 = np.array(['34','7','9','10'], dtype=np.uint8)
print(a4)
print(a4.dtype)
```

```python
# Forzar conversión a otro tipo
a5 = a4.astype(np.int16)
a5.dtype
```

```python
a6 = a5.astype('u4')
a6.dtype
```

```python
arr = np.array([1,2,3,5,4,33,2,1,2,66,754,9])
v = arr > 10
v
```

```python
v2 = arr[arr > 10]
v2
```

```python
# Filtrar mediante un intervalo: & and, | or
v3 = arr[(arr > 2) & (arr < 10)]
v3
```
