```python
import matplotlib.pyplot as plt
np.random.seed(1)
# 1000 random integers between 0 and 50
x = np.random.randint (0, 50,1000)
y = x + np.random.normal (0, 10, 1000)
plt.scatter(x,y, color="blue")
c = np.corrcoef (x, y)
plt.show()
```

```python
c
```

```python
round(c[0][1] * 100, 2)
```
